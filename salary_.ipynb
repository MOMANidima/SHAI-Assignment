```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
df.columns
```

```python
print(df.info())
```

```python
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode().iloc[0]  # Mode returns a Series, so we use iloc[0] to get the first mode
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()

print ('mean = ' + str(mean_salary) )
print ('median = ' + str(median_salary))
print ('mode = ' + str(mode_salary))
print ('minimum = ' + str(min_salary))
print ('maximum = ' + str(max_salary))
```

```python
df.std()
```

```python
from matplotlib import pyplot as plt
import numpy as np
data = df.JobTitle
data_series = pd.Series(data)

# Count occurrences of each unique item in the data
counts = data_series.groupby(data_series).size()

# Create a pie chart
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Job Title')

#plt.figure(figsize=(1000000, 1000000))

# Display the chart
plt.show()
```
